```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sol(M, h, T, n):
    x_av = np.zeros(int(T / h) + 1)
    t = []
    for i in range(int(T / h) + 1):
        t.append(i * h)
    for l in range(M):
        a = np.zeros((2 * n + 1, 2 * n + 1, 2 * n + 1))
        b = np.random.choice(2, size = (2 * n + 1, 2 * n + 1, 2 * n + 1), p = [0.99, 0.01])
        for i in range(2 * n + 1):
            for j in range(2 * n + 1):
                for k in range(2 * n + 1):
                    if ((i - n) ** 2 + (j - n) ** 2 + (k - n) ** 2) != 0 and b[i][j][k]:
                        a[i][j][k] = (1 + 3 * (n - k) ** 2 / ((n - i) ** 2 + (n - j) ** 2 + (n - k) ** 2)) ** 0.5 / (((n - i) ** 2 + (n - j) ** 2 + (n - k) ** 2)) ** 1.5
        x = np.array([])
        for i in range(int(T / h) + 1):
            x = np.append(x, np.prod(np.cos(a / 2 * (i * h))))
        x_av += x / M
    return t, x_av
```

```python
t, x = sol(1, 0.01, 40, 20)
plt.plot(t, x)
plt.show()
```
